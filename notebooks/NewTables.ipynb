```python
import geopandas as gpd
from sqlalchemy import create_engine

# Conexión a PostGIS
engine = create_engine('postgresql://postgres:MelonSK998@localhost:5432/Poligonos')

# Función para cargar tablas desde PostGIS
def cargar_tabla(nombre_tabla, gcol):
    query = f"SELECT * FROM {nombre_tabla};"
    gdf = gpd.read_postgis(query, engine, geom_col=f"{gcol}")
    return gdf
```

```python
gdf_poligonos_manzana = cargar_tabla("poligonos_manzanas_agebs_colonias", gcol="GEOM_MANZANA" )
```

```python
gdf_demograficos = gpd.read_postgis("SELECT * FROM datos_demograficos_particionada;",
        engine, geom_col=f"geometry")
```

```python
poligonos_demograficos = pd.merge(gdf_poligonos_manzana, 
                    gdf_demograficos, 
                    left_on= 'ID_AGEB',right_on='ageb', how = 'left')
```
